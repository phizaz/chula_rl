```python
def show_board(board):
    print("Parts:", *np.diag(board[2]))
    img = np.ones(board.shape[1:] + (3,))
    for i in range(3):
        img[:, :, i] = 1.0
    current = board[1] == -1
    img[:, :, 0] -= 0.6 * current
    img[:, :, 1] -= 0.6 * current
    current = board[1] == -2
    img[:, :, 0] -= 0.3 * current
    img[:, :, 1] -= 0.3 * current
    opponent = board[1] == 1
    img[:, :, 1] -= 0.6 * opponent
    img[:, :, 2] -= 0.6 * opponent
    opponent = board[1] == 2
    img[:, :, 1] -= 0.3 * opponent
    img[:, :, 2] -= 0.3 * opponent
    
    fig, ax = plt.subplots(1, 1, figsize=(10, 10))
    ax.imshow(img)
    r, c = board[0].shape
    for i in range(r):
        for j in range(c):
            text = ax.text(j, i, board[0,i,j], fontsize='40', ha='center', va='center')
            
    plt.show()
```
